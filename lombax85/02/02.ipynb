```python
import re


f = open("./input.txt", "r")
lines = f.readlines()

class Game:
    id = 0
    steps = []

    def __init__(self, id, steps):
        self.id = id
        self.steps = steps
    
    # sol 1
    def is_game_possible(self, red, green, blue):
        totals = []
        for step in self.steps:
            for stepitem in step:
                color = stepitem[1]
                number_of_items = stepitem[0]
                if color == "red":
                    if number_of_items > red: return False
                if color == "green":
                    if number_of_items > green: return False
                if color == "blue":
                    if number_of_items > blue: return False
        return True
    
    # sol 2
    def minimum_cubes(self):
        totals = []
        min_cubes = {
            'red' : 0,
            'green' : 0,
            'blue' : 0
        }
        for step in self.steps:
            for stepitem in step:
                color = stepitem[1]
                number_of_items = stepitem[0]
                if color == "red":
                    if number_of_items > min_cubes['red']:
                        min_cubes['red'] = number_of_items
                if color == "green":
                    if number_of_items > min_cubes['green']:
                        min_cubes['green'] = number_of_items
                if color == "blue":
                    if number_of_items > min_cubes['blue']:
                        min_cubes['blue'] = number_of_items

        return min_cubes['red']*min_cubes['green']*min_cubes['blue']



def parse_game_string(game_string):
    pattern = r"Game (\d+): (.*)"
    match = re.match(pattern, game_string)
    if not match:
        return None
    
    game_id, steps_str = match.groups()
    steps_list = steps_str.split("; ")

    steps = []
    for step in steps_list:
        pairs = step.split(", ")
        step_data = []
        for pair in pairs:
            number, color = pair.split(" ")
            step_data.append([int(number), color])
        steps.append(step_data)

    return {"game_id": int(game_id), "steps": steps}


games = []
games_possible_1 = []
minimum_cubes_2 = []
for line in lines:
    res = parse_game_string(line)
    game = Game(res['game_id'], res['steps'])
    games.append(game)
    if game.is_game_possible(12,13,14) == True:
        games_possible_1.append(game.id)
    minimum_cubes_2.append(game.minimum_cubes())

print("Solution 1: ", sum(games_possible_1))
print("Solution 2: ", sum(minimum_cubes_2))

```
